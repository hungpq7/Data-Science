# 1. Matplotlib API
Matplotlib is a famous visualization library in Python, is the foundation for many other visualization libraries. Matplotlib along with NumPy can be considered equivalent to Matlab.

Since Matplotlib is a very large library, this topic focuses only on Pyplot, the most essential sub-module of Matplotlib.

## 1.1. Plotting interfaces

#### Object-oriented interface
Every plot created by Matplotlib is under the the control of two objects, `figure` and `axes`.
- A `figure` object is the whole image generated by Matplotlib. it works as a container for a number of `axes` objects.
- Each `axes` object (a pair of *axes*), usually refers to a 2-dimensional Cartesian coordinate system. It gives users access to plot elements.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style('whitegrid')
```

```python
# create a figure
fig = plt.figure()

# add an axes with some text
ax = fig.add_subplot()
ax.text(0.5, 0.5, 'DATA', ha='center', va='center')

plt.show()
```

```python
x1 = np.linspace(-3, 3, 100)
y1 = x1**2
x2 = np.linspace(-3, 3, 100)
y2 = np.sin(x2)
```

```python
# create a figure and a single axes
fig, ax = plt.subplots()
ax.plot(x1, y1)
plt.show()
```

## 1.2. Figure and axes
The `subplots()` function of Matplotlib creates two objects, a figure and a matrix of axes. Here are some notable parameters:
- The `figsize` parameter allow you to change the figure width and height (in inches).
- The `nrows` and `ncols` parameters indicate the shape of the matrix of axes.
- `sharex` controls either subplots are using the same $x$-axis or not. So does `sharey`.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-deep')
sns.set_style('whitegrid')
```

```python
fig, ax = plt.subplots()
ax.axis('scaled')
plt.show()
```

Now create a $1\times2$ matrix of axes. You can use slicing to access each axes.

```python
x = np.arange(-100, 100)
y = x**2
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
ax[1].plot(x, y, linewidth=3, color='indianred')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True)
ax[1, 0].plot(x, y, linewidth=3)
plt.show()
```

```python
x1 = np.linspace(-5, 5, 100)
y1 = x**2-1
x2 = np.linspace(-5, 5, 100)
y2 = 1/4*x+2

fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(x2, y2)
ax.axis('scaled')
ax.grid(linestyle=':')
# ax.set_xticks(np.arange(-5, 6, 1))
ax.legend(['a', 'b'], loc='best')
ax.set_frame_on(False)
ax.tick_params(bottom=False, left=False)
ax.axvline(x=0, lw=1, color='k')
ax.axhline(y=0, lw=1, color='k')
plt.show()
```
